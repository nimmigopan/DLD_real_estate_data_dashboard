```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

website = 'https://dubailand.gov.ae/en/open-data/real-estate-data/#/'
driver = webdriver.Chrome()
driver.get(website)

# Clicking on the calendar
driver.find_element(By.ID, "transaction_pFromDate").click()
time.sleep(1)
driver.find_element(By.XPATH, '//*[@id="ui-datepicker-div"]/div/a[1]/span').click()  # navigating to Jan
time.sleep(1)
driver.find_element(By.XPATH, '//*[@id="ui-datepicker-div"]/table/tbody/tr[1]/td[2]/a').click()  # clicking on 1
time.sleep(1)

# Clicking on the search button
search_button = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="trxFilter"]/div/div[10]/div/button[1]')))
search_button.click()

while True:
    # Re-locate the table after each click on the next button
    table = WebDriverWait(driver, 10).until(EC.visibility_of_element_located((By.ID, "transactionGrid")))
    rows = table.find_elements(By.TAG_NAME, "tr")
    
    for row in rows:
        cells = row.find_elements(By.TAG_NAME, "td")
        row_data = [cell.text for cell in cells]
        print(row_data)
        
    next_button = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="transactionGrid_next"]')))
    if "disabled" in next_button.get_attribute("class"):
        break

    # Scroll to the next button
    driver.execute_script("arguments[0].scrollIntoView();", next_button)

    # Click on the next button using JavaScript
    driver.execute_script("arguments[0].click();", next_button)

driver.quit()
```
